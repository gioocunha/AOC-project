```python
# Gráfico gerado para o relatório de AOC

import matplotlib.pyplot as plt
import numpy as np

# Estatísticas de tempo de execução
tempo_referencia = [1.0890, 1.3733, 2.7170, 0.1342]
tempo_sistema_1 = [0.9070, 1.0547, 0.8444, 0.1595]
tempo_sistema_2 = [0.7108, 0.8825, 0.6812, 0.0850]
tempo_sistema_3 = [3.1319, 3.8653, 2.5167, 0.3709]

programas = ['Programa 1', 'Programa 2', 'Programa 3', 'Programa 4']

# largura das barras e a posição dos grupos
largura_barra = 0.2
posiçao_ref = np.arange(len(programas))
posiçao_sistema_1 = [x + largura_barra for x in posiçao_ref]
posiçao_sistema_2 = [x + largura_barra for x in posiçao_sistema_1]
posiçao_sistema_3 = [x + largura_barra for x in posiçao_sistema_2]
posiçao_sistema_4 = [x + largura_barra for x in posiçao_sistema_3]

# Criando o gráfico de barras
plt.bar(posiçao_ref, tempo_referencia, largura_barra, label='Sistema Referência')
plt.bar(posiçao_sistema_1, tempo_sistema_1, largura_barra, label='Sistema 1')
plt.bar(posiçao_sistema_2, tempo_sistema_2, largura_barra, label='Sistema 2')
plt.bar(posiçao_sistema_3, tempo_sistema_3, largura_barra, label='Sistema 3')

        
# Definindo as legendas dos eixos e do gráfico
plt.xlabel('Programas')
plt.ylabel('Tempo de Execução (s)')
plt.title('Tempo de Execução de Cada Programa nos Sistemas')
plt.xticks(posicao_sistema_1 + largura_barra, programas)
plt.legend()
```
